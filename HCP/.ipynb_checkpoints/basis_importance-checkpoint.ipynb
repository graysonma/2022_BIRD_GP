```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import pickle
import random
import sys
import seaborn as sns

sys.path.append(",,/birdgp")
import utils

%matplotlib inline
```

```python
outcome_name = "socirand"
mode_list = ["fALFF+confounder", "connectivity+confounder", "fALFF+connectivity+confounder"]

basis_importance = []
Psi = []
for mode_idx in range(3):
    mode = mode_list[mode_idx]
    
    if mode_idx == 0:
        with open("/scratch/jiankang_root/jiankang0/gxma/IIR/jasa_results_connectivity+confounder/{}_importance/reduced1_model_{}.pickle".format(outcome_name, mode), "rb") as f:
            model = pickle.load(f)
    else:
        with open("/scratch/jiankang_root/jiankang0/gxma/IIR/jasa_results_connectivity+confounder/{}_importance/model_{}.pickle".format(outcome_name, mode), "rb") as f:
            model = pickle.load(f)

    grad_basis = utils.basis_importance(model)

    mean_grad_basis = np.mean(grad_basis[:, :, 1:151], 0)
    mean_grad_basis = np.mean(mean_grad_basis, 0)
    abs_mean_grad_basis = abs(mean_grad_basis)
    basis_importance.append(abs_mean_grad_basis)
```

```python
order_fALFF = np.flip(np.argsort(basis_importance[0]))
order_connectivity = np.flip(np.argsort(basis_importance[1]))
order_both = np.flip(np.argsort(basis_importance[2]))
```

```python
Psi_fALFF = np.load("/scratch/jiankang_root/jiankang0/gxma/IIR/jasa_results_connectivity+confounder/fit_basis/fALFF_Psi_all_for_importance.npy")
Psi_connectivity = np.load("/scratch/jiankang_root/jiankang0/gxma/IIR/jasa_results_connectivity+confounder/fit_basis/connectivity_Psi_all_for_importance.npy")
Psi_both = np.vstack((Psi_fALFF, Psi_connectivity))
Psi_fALFF, _, _ = np.linalg.svd(Psi_fALFF, full_matrices = False)
Psi_connectivity, _, _ = np.linalg.svd(Psi_connectivity, full_matrices = False)
Psi_both, _, _ = np.linalg.svd(Psi_both, full_matrices = False)
```

```python
top_Psi_fALFF = Psi_fALFF[:, order_fALFF]
top_Psi_connectivity = Psi_connectivity[:, order_connectivity]
top_Psi_fALFF_from_both = Psi_both[:Psi_fALFF.shape[0], order_both]
top_Psi_connectivity_from_both = Psi_both[Psi_fALFF.shape[0]:, order_both]
```

```python
import pandas as pd
block = pd.read_csv("PowerParcellation_coordinates_networklabels.csv")
block = np.array(block["network_id"])
order = np.argsort(block)
unique_idx = np.sort(np.unique(block))
count = []
for idx in range(len(unique_idx)):
    count.append(np.sum(block == unique_idx[idx]))
groups = np.concatenate(([0], np.cumsum(count)))
group_names = ["UC", "SMH", "SMM", "CO", "AD", "DM", "MR", "VS", "FP", "SA", "SC", "VAT", "DAT", "CRB"]
group_number = len(group_names)
width = 10
d = np.diff(groups)
ticks = groups[:-1] + d / 2
```

```python
import matplotlib.patches as patches

plot_basis_idx = [0, 1, 2, 4, 101, 103]
plot_num = len(plot_basis_idx)
for plot_idx in range(plot_num):
    connectivity = np.zeros((264, 264))
    idx = np.triu_indices(264, k = 1)
    connectivity[idx] = Psi_both[Psi_fALFF.shape[0]:, plot_basis_idx[plot_idx]]
    connectivity = connectivity + connectivity.T
    connectivity = connectivity[order, :][:, order]
    connectivity[connectivity == 0] = np.nan
    max_val = np.nanmax(connectivity)
    min_val = np.nanmin(connectivity)
    scaled_connectivity = 2 * (connectivity - min_val) / (max_val - min_val) - 1
    padded = np.vstack((np.tile(np.nan, (width, 264)), scaled_connectivity))
    padded = np.hstack((padded, np.tile(np.nan, (264 + width, width))))
    
    fig, ax = plt.subplots(figsize = (4, 4), dpi = 200)   
    sns.heatmap(padded, 
                cbar = False
                #cbar_kws = {"orientation": "horizontal"}
               )
    plt.xticks(ticks = ticks, labels = group_names, rotation = 90, size = 6)
    plt.yticks(ticks = ticks + width, labels = group_names, rotation = 0, size = 6)
    ax.xaxis.set_ticks_position('top')
    ax.yaxis.set_ticks_position('right')
    ax.tick_params(axis = 'both', which = 'both', length = 0)
    for g in range(group_number):
        rect = patches.Rectangle((groups[g], 0), d[g], 10, linewidth = 0, facecolor = ['darkgrey', 'lightgrey'][g % 2])
        ax.add_patch(rect)
        rect = patches.Rectangle((264, 10 + groups[g]), 10, d[g], linewidth = 0, facecolor = ['darkgrey', 'lightgrey'][g % 2])
        ax.add_patch(rect)
        if g < (group_number - 1):
            ax.vlines(groups[g + 1], ymin = width, ymax = groups[g + 1] + width, color = 'darkgrey', linewidth = 0.5)
            ax.hlines(groups[g + 1] + width, xmin = groups[g + 1], xmax = 264, color = 'darkgrey', linewidth = 0.5)
    plt.savefig("basis_images/connectivity_{}.png".format(plot_idx));
```
