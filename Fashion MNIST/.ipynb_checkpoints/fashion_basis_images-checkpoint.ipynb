```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Psi_predictors = np.loadtxt("Psi_predictors.txt", delimiter = ",")
Psi_outcomes = np.loadtxt("Psi_outcomes.txt", delimiter = ",")
theta_train_predictors = np.loadtxt("theta_train_predictors.txt", delimiter = " ")
theta_test_predictors = np.loadtxt("theta_test_predictors.txt", delimiter = " ")
theta_train_outcomes = np.loadtxt("theta_train_outcomes.txt", delimiter = " ")
theta_test_outcomes = np.loadtxt("theta_test_outcomes.txt", delimiter = " ")
```

```python
Psi_predictors.shape
```

### Predictor Basis

```python
fg, ax = plt.subplots(3, 2, figsize = (10, 4.5))
vmin = np.min(Psi_predictors[:, :6])
vmax = np.max(Psi_predictors[:, :6])

idx = 0
for r in range(3):
    for c in range(2):
        ax[r, c].matshow(Psi_predictors[:, idx].reshape((28, 112)), cmap = "gray", vmin = vmin, vmax = vmax)
        idx = idx + 1
        
        ax[r, c].set_xticks([])
        ax[r, c].set_yticks([])

plt.tight_layout()
plt.savefig("basis_images_predictors.png")
plt.show()
```

```python
# training predictor
fg, ax = plt.subplots(1, 2, figsize = (10, 4))
ax[0].plot(np.sort(np.mean(abs(theta_train_predictors), 0))[::-1])
ax[1].plot(np.sort(np.std(theta_train_predictors, 0))[::-1])
plt.show()
```

```python
plt.plot(np.cumsum(np.sort(np.mean(abs(theta_train_predictors), 0))[::-1] / sum(np.sort(np.mean(abs(theta_train_predictors), 0))[::-1])))
```

```python
np.cumsum(np.sort(np.std(theta_train_predictors, 0))[::-1] / sum(np.sort(np.std(theta_train_predictors, 0))[::-1]))
```

```python
# training predictor
fg, ax = plt.subplots(1, 2, figsize = (10, 4))
ax[0].plot(np.sort(np.mean(abs(theta_test_predictors), 0))[::-1])
ax[1].plot(np.sort(np.std(theta_test_predictors, 0))[::-1])
plt.show()
```

### Outcome Basis

```python
fg, ax = plt.subplots(3, 2, figsize = (4, 6))
vmin = np.min(Psi_outcomes[:, :6])
vmax = np.max(Psi_outcomes[:, :6])

idx = 0
for r in range(3):
    for c in range(2):
        ax[r, c].matshow(Psi_outcomes[:, idx].reshape((28, 28)), cmap = "gray", vmin = vmin, vmax = vmax)
        idx = idx + 1
        
        ax[r, c].set_xticks([])
        ax[r, c].set_yticks([])

plt.tight_layout()
plt.savefig("basis_images_outcomes.png")
plt.show()
```

```python
# training outcome coefficients
fg, ax = plt.subplots(1, 2, figsize = (10, 4))
ax[0].plot(np.sort(np.mean(abs(theta_train_outcomes), 0))[::-1])
ax[1].plot(np.sort(np.std(theta_train_outcomes, 0))[::-1])
plt.show()
```

```python
np.cumsum(np.sort(np.mean(abs(theta_train_outcomes), 0))[::-1] / sum(np.sort(np.mean(abs(theta_train_outcomes), 0))[::-1]))
```

```python
np.cumsum(np.sort(np.std(theta_train_outcomes, 0))[::-1] / sum(np.sort(np.std(theta_train_outcomes, 0))[::-1]))
```

```python
# testing outcome coefficients
fg, ax = plt.subplots(1, 2, figsize = (10, 4))
ax[0].plot(np.mean(abs(theta_test_outcomes), 0))
ax[1].plot(np.std(theta_test_outcomes, 0))
plt.show()
```

```python
# testing outcome coefficients
total_predictors = np.sum(np.sort(np.var(theta_test_predictors, 0))[::-1])
total_outcomes = np.sum(np.sort(np.var(theta_test_outcomes, 0))[::-1])

fg, ax = plt.subplots(1, 2, figsize = (10, 4))
ax[0].plot(np.sort(np.var(theta_test_predictors, 0))[::-1] / total_predictors)
ax[0].set_ylabel('variantion explained')
ax[0].set_xlabel('basis')
ax[1].plot(np.sort(np.var(theta_test_outcomes, 0))[::-1] / total_outcomes)
ax[1].set_ylabel('variantion explained')
ax[1].set_xlabel('basis')
plt.show()
```

```python
plt.figure(figsize = (5, 4))
plt.plot(np.cumsum(np.sort(np.var(theta_test_predictors, 0))[::-1] / total_predictors))
plt.ylabel('proportion of variance explained')
plt.ylim([0.15, 1.05])
plt.tight_layout()
plt.savefig("variation_proportion_1.png")
```

```python
plt.figure(figsize = (5, 4))
plt.plot(np.cumsum(np.sort(np.var(theta_test_outcomes, 0))[::-1] / total_outcomes))
plt.ylabel('proportion of variance explained')
plt.ylim([0.15, 1.05])
plt.tight_layout()
plt.savefig("variation_proportion_2.png")
```
