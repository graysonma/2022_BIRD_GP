## BIRD-GP: A Fashion MNIST Example

### Install from Github

```python
import sys
sys.path.append("../birdgp")

import numpy as np
import matplotlib.pyplot as plt
import math
import bird_gp
import random
import torch

%matplotlib inline
```

### Read images

```python
import zipfile
with zipfile.ZipFile("images.zip", 'r') as zip_ref:
    zip_ref.extractall("")
train_predictors = np.loadtxt("train_predictors.txt")
test_predictors = np.loadtxt("test_predictors.txt")
train_outcomes = np.loadtxt("train_outcomes.txt")
test_outcomes = np.loadtxt("test_outcomes.txt")
```

### Some example images

```python
i1, i2, i3, i4 = 0, 1, 2, 3
fg, ax = plt.subplots(4, 2, figsize = (12, 7))
ax[0, 0].matshow(train_predictors[i1, :].reshape((28, 112)), cmap = "gray")
ax[1, 0].matshow(train_predictors[i2, :].reshape((28, 112)), cmap = "gray")
ax[2, 0].matshow(train_predictors[i3, :].reshape((28, 112)), cmap = "gray")
ax[3, 0].matshow(train_predictors[i4, :].reshape((28, 112)), cmap = "gray")

ax[0, 1].matshow(train_outcomes[i1, :].reshape((28, 28)), cmap = "gray")
ax[1, 1].matshow(train_outcomes[i2, :].reshape((28, 28)), cmap = "gray")
ax[2, 1].matshow(train_outcomes[i3, :].reshape((28, 28)), cmap = "gray")
ax[3, 1].matshow(train_outcomes[i4, :].reshape((28, 28)), cmap = "gray")

for i in range(4):
    for j in range(2):
        ax[i, j].set_xticks([])
        ax[i, j].set_yticks([])

plt.tight_layout()
plt.show()
```

### Generate grids

```python
predictor_grids = bird_gp.generate_grids([28, 112])
outcome_grids = bird_gp.generate_grids([28, 28])
```

### Create BIRD_GP object and fit

```python
birdgp = bird_gp.BIRD_GP(predictor_grids = predictor_grids,
                         outcome_grids = outcome_grids,
                         predictor_L = 50,
                         outcome_L = 50,
                         svgd_b_lambda = 1e2, 
                         bf_predictor_steps = 10000,
                         bf_outcome_steps = 10000)
birdgp.fit(train_predictors, train_outcomes)
```

### Evaluation on training data

```python
train_pred = birdgp.predict_train()
np.mean((train_pred - train_outcomes)**2)
```

### Evaluation on testing data

```python
test_pred = birdgp.predict_test(test_predictors)
np.mean((test_pred - test_outcomes)**2)
```

### Plot basis images

```python
# first 6 basis images of predictors
fg, ax = plt.subplots(3, 2, figsize = (10, 4.5))
vmin = np.min(birdgp.Psi_predictors[:, :6])
vmax = np.max(birdgp.Psi_predictors[:, :6])

idx = 0
for r in range(3):
    for c in range(2):
        ax[r, c].matshow(birdgp.Psi_predictors[:, idx].reshape((28, 112)), cmap = "gray", vmin = vmin, vmax = vmax)
        idx = idx + 1
        
        ax[r, c].set_xticks([])
        ax[r, c].set_yticks([])

plt.tight_layout()
plt.show()
```

```python
# first 6 basis images of outcomes
fg, ax = plt.subplots(3, 2, figsize = (4, 6))
vmin = np.min(birdgp.Psi_outcomes[:, :6])
vmax = np.max(birdgp.Psi_outcomes[:, :6])

idx = 0
for r in range(3):
    for c in range(2):
        ax[r, c].matshow(birdgp.Psi_outcomes[:, idx].reshape((28, 28)), cmap = "gray", vmin = vmin, vmax = vmax)
        idx = idx + 1
        
        ax[r, c].set_xticks([])
        ax[r, c].set_yticks([])

plt.tight_layout()
plt.show()
```

### Proportion of variation explained

```python
# predictors
predictors_total = np.sum(np.var(birdgp.theta_train_predictors, 0))

plt.figure(figsize = (5, 4))
plt.plot(np.cumsum(np.sort(np.var(birdgp.theta_train_predictors, 0))[::-1] / predictors_total))
plt.ylabel('proportion of variance explained')
plt.ylim([0.15, 1.05])
plt.tight_layout()
```

```python
# outcomes
outcomes_total = np.sum(np.var(birdgp.theta_train_outcomes, 0))

plt.figure(figsize = (5, 4))
plt.plot(np.cumsum(np.sort(np.var(birdgp.theta_train_outcomes, 0))[::-1] / outcomes_total))
plt.ylabel('proportion of variance explained')
plt.ylim([0.15, 1.05])
plt.tight_layout()
```
